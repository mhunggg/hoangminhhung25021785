```python
#W6A4
a = input().split()
result = {}
for i in a:
    key, value = i.split(':')
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
print(result)
```

```python
#W6A10
a=list(map(int,input().split()))
b=int(input())
dic = {}
lst = []
for i in range(len(a)):
    if b-a[i] in dic:
        lst.append((dic[b-a[i]],a[i]))
    dic[a[i]]=a[i]
print(sorted(lst))
```

```python
#W6A12 10 20 5 30 15
a=list(map(int,input().split()))
dic = {}
max = 0
test = []
for i in a:
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1
for j in a:
    if dic[j]>max:
        max=dic[j]
        test = j
    elif dic[j]==max and j<test:
        max=dic[j]
        test=j
print(test)

```

```python
#W6A13
a=list(map(str,input().split()))
value=[]
key=[]
dict = {}
for i in range(0,len(a)):
    if i%2==0:
        value.append(a[i])
    else:
        key.append(a[i])
for x,y in zip(key,value):
    dict[x]=y
print(dict)
```

```python
#W6A14
a=list(map(str,input().split()))
b=list(map(str,input().split()))
lst = []
for i in a:
    for j in b:
        if i==j and int(j) not in lst:
            lst.append(int(i))
print(lst)
```
