```python
#W5A5
a=input().split()
b=input()
def hello(a,b):
    x = 0
    check = False
    for i in a:
        x += 1
        if i == b:
            check = True
            break
    if check == True:
        return x
    else:
        return -1
print(hello(a,b))
```

```python
#W5A6
a=int(input())
def factorial(a):
    b=1
    for i in range(1,a+1):
        b*=i
    return b
print(factorial(a))
```

```python
#W5A7
a=input().split()
def calc(a):
    b = float(a[0])
    c = float(a[2])
    if a[1] == "+":
        print("{:.2f}".format(b+c))
    elif a[1] == "-":
        print("{:.2f}".format(b-c))
    elif a[1] == "*":
        print("{:.2f}".format(b*c))
    elif a[1] == "/":
        print("{:.2f}".format(b/c))
calc(a)
```

```python
#W5A8
a,b=map(int,input().split())
def hamming(a,b):
    m = bin(a^b)[2:]
    sum = 0
    for i in m:
        sum += int(i)
    return sum
print(hamming(a,b))
```

```python
#W5A10
a,b=map(str,input().split())
def isomorphism(a,b):
    for i in range(len(a)):
        for j in range(len(a)):
            if (a[i]==a[j] and b[i]!=b[j]) or (b[i]==b[j] and a[i]!=a[j]):
                return False
    return True
print(isomorphism(a,b))
```
