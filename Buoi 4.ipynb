```python
n = int(input())
lst1 = []
lst2 = []
for i in range(2,n+1):
    if n%i == 0:
        lst1.append(i)
for x in lst1:
    if x <= 1 :
        continue
    elif x == 2:
        lst2.append(2)
    else:
        check = True
        for y in range(2,x+1):
            if x % y == 0  and y != x:
                check = False
                break
        if check:
            lst2.append(x)
print(max(lst2))
```
