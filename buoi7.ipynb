```python
#W7A1 1 2 2 3 3 4 5 6 7 8 | 9
a = sorted(list(map(int,input().split())))
b = int(input())
left = 0
right = len(a) - 1
def binary_search(a,left,right,b):
    mid = (left+right)//2
    if left <= right:
        if a[mid] < b:
            return binary_search(a,mid+1,right,b)
        elif a[mid] > b:
            return binary_search(a,left,mid-1,b)
        elif a[mid]==b:
            return mid
    else:
        return -1
print(binary_search(a,left,right,b))
```

```python
#W7A3 3 2 4 5 7 8 6 1 2 3
a = list(map(int,input().split()))
def bubble_sort(a):
    n = len(a)
    for i in range(n):
        swapped = False
        for j in range(0,n-i-1):
            if a[j] > a[j+1]:
                a[j] , a[j+1] = a[j+1] , a[j]
                swapped = True
        if not swapped:
            break
    return a
print(bubble_sort(a))
```

```python
#W7A10
import bisect
nums = eval(input())
m = int(input())
def max_subarray_mod(nums, m):
    max_mod = 0
    prefix_sum = 0
    sorted_prefixes = [0] 
    
    for num in nums:
        prefix_sum = (prefix_sum + num) % m
        max_mod = max(max_mod, prefix_sum)
        idx = bisect.bisect_right(sorted_prefixes, prefix_sum)
        
        if idx < len(sorted_prefixes):
            current_mod = (prefix_sum - sorted_prefixes[idx] + m) % m
            max_mod = max(max_mod, current_mod)
        bisect.insort(sorted_prefixes, prefix_sum)
        
    return max_mod
print(max_subarray_mod(nums,m))
```
