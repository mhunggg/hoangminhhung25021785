```python
#W11A3
def happy_number(n):
    hello = ()
    while n != 1:
        if n in hello:
            return False
        
        hello.append(n)

        n = sum(int(i) ** 2 for i in str(n))
    return True

try:
    n = int(input())
    if happy_number(n):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("NO")
```

```python
#W11A10
def valid(a):
    seen = set()
    cur = 0
    for i in a:
        if i in seen:
            return 'Mang khong hop le'
        seen.add(i)
        cur += i
    return cur

a = list(map(int, input().split()))
print(valid(a))
```

```python
#W11A11
import re

def check_regex():
    try:
        t_str = input().strip()
        if not t_str:
            return
        t = int(t_str)
        
        for _ in range(t):
            s = input().strip()
            try:
                re.compile(s)
                print("True")
            except re.error:
                print("False")
                
    except ValueError:
        print("Vui lòng nhập một số nguyên cho T.")

check_regex()
```
